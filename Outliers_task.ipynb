#**Outlier Detection**
##### This is where you have to code. You have 3 parts in this assignment. The first two parts are compulsory. The third part is optional.

## Method 1 - Bollinger Bands

```python
## Your Code Goes Here
values_arr = list(DataDict.values())
bittu = values_arr[0]
mean = [0]*10
sd = [0]*10
n = 10
for i in range(10, len(bittu)):
    mean.append((sum(bittu[i-10:i-1]))/n)

for i in range(10, len(bittu)):
    t = 0
    for i in range(i-10, i):
        t += pow(bittu[i] - mean[i], 2)
    me = (t/(n-1))**(1/2)
    sd.append(me)

k = 2.5
anomaly_times = []
for i in range(10, len(bittu)):
    x = bittu[i]
    if x > mean[i] + k * sd[i] or x < mean[i] - k * sd[i]:
        anomaly_times.append(i)
```

## Method 2 - Slope Distribution

```python
## Your Code Goes Here
slope = []
res = []
value_arr = list(DataDict.values())
a = value_arr[0]
for i in range(1, len(a)):
    slope.append(a[i] - a[i-1])

mean = (sum(slope))/(len(slope))
p = 0
for i in range(len(slope)):
    p += pow(slope[i] - mean, 2)
sd = (p/(len(slope)-1))**(1/2)
for i in range(len(a)):
    x = a[i]
    if x < mean - 3.5 * sd or x > mean + 3.5 * sd:
        res.append(i)
```
